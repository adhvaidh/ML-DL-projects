```python
ds=pd.read_csv('Mall_Customers.csv')
```

```python
ds# Here there's no dependent variable, therefore unsupervised ML
```

```python
x=ds.iloc[:,[3,4]].values
```

```python
x
```

```python
#Elbow meathod :
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):  #10 K-value
    kmeans=KMeans(n_clusters=i, init='k-means++')  #init - iteration; k-means++ means centroid.
    kmeans.fit(x)
    wcss.append(kmeans.inertia_) #intertia_ means euclidean method value
import matplotlib.pyplot as plt 
plt.plot (range(1,11), wcss)
```

```python
wcss
```

```python
kmeans=KMeans(n_clusters=5, init='k-means++')
y_kmeans=kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
pd.concat([ds, pd.DataFrame(y_kmeans)], axis=1)
```

```python
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1], s=500, c='red', label='Cluster1')
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1], s=500, c='pink', label='Cluster2')
plt.scatter(x[y_kmeans==2,0], x[y_kmeans==2,1], s=500, c='yellow', label='Cluster3')
plt.scatter(x[y_kmeans==3,0], x[y_kmeans==3,1], s=500, c='black', label='Cluster4')
plt.scatter(x[y_kmeans==4,0], x[y_kmeans==4,1], s=500, c='green', label='Cluster5')
```
